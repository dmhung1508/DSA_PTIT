```python
def phanloaii1(text):
  completion = openai.ChatCompletion.create(
  model="ft:gpt-3.5-turbo-0613:personal::7yLjTnqy",
    messages=[
      {"role": "system", "content": "Please classify the emotions for the following sentence into 3 types of negative, positive or neutral, each answer only one type"},
      {"role": "user", "content": text}
    ]
  )
  a = completion.choices[0].message
  print(a["content"])
  return a["content"]
```
