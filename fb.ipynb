```python
import threading
import time
# Số lượng luồng bạn muốn tạo
num_threads = 10
while True:
    # Danh sách các luồng
    threads = []

    # Tạo và khởi chạy các luồng
    for _ in range(num_threads):
        thread = threading.Thread(target=chat)
        thread.start()
        threads.append(thread)

    # Đợi tất cả các luồng kết thúc
    for thread in threads:
        thread.join()

    # Tiếp tục thực hiện các công việc khác sau khi tất cả các luồng đã kết thúc
    print("All chat threads have finished.")
```

```python
from multiprocessing import Pool

def chat_function(thread_number):
    # Hàm chat_function() chứa mã logic cho việc chạy chat

    # Code của bạn ở đây

    # Ví dụ: In ra một số thông điệp chat với số hiệu của thread
    for i in range(5):
        print("Chat message from process", thread_number, ":", i)

# Tạo một Pool với số lượng tiến trình là 100
pool = Pool(processes=100)

# Khởi chạy các tiến trình bằng cách áp dụng công việc cho pool
pool.map(chat_function, range(100))

# Đợi tất cả các tiến trình kết thúc
pool.close()
pool.join()

# Tiếp tục thực hiện các công việc khác sau khi tất cả các tiến trình đã kết thúc
print("All chat processes have finished.")
```
